```python
%%cython
cimport cython
from libc.stdio cimport fopen, FILE, fclose, fprintf

def c_write_to_file(filename, content):
    filename_byte_string = filename.encode("UTF-8")
    cdef char* fname = filename_byte_string
    cdef char* line = content
    
    cdef FILE* cfile
    cfile = fopen(fname, "w")
    if cfile == NULL:
        return
    
    fprintf(cfile, line)
    fclose(cfile)
    
    return []
```

```python
def python_write_to_file(filename, content):
    with open(filename, 'w') as f:
        f.write(content)
```

```python
%%cython
cimport cython
from libc.stdio cimport fopen, FILE, fclose, fprintf

def write_data_cython(object filename, list data):
    filename_byte_string = filename.encode("UTF-8")
    cdef char* fname= filename_byte_string
    cdef FILE* cfile
    cdef float a, b, c
    
    cfile = fopen(fname, "wb")
    if cfile == NULL:
        return
    
    cdef char* fmt = "%f,%f,%f\n"
    for i in range(10000):
        a, b, c = data
        fprintf(cfile ,fmt, a,b,c)

    fclose(cfile)
```
