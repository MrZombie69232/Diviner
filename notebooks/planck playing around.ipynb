```python
import physics
```

```python
waves = arange(1e-6,400e-6,1e-6)
```

```python
radiance = physics.planck_wavelength(60,waves)
```

```python
plot(waves*1e6,radiance)
```

```python
import pandas as pd
```

```python
radiances = pd.DataFrame(index=waves*1e6)
```

```python
radiances['60']=radiance
```

```python
radiances['50']=physics.planck_wavelength(50,waves)
```

```python
radiances['20']=physics.planck_wavelength(20,waves)
```

```python
radiances[['20','30','40']].plot()
```
