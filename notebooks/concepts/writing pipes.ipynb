```
from diviner import file_utils as fu
```

```
def cols_to_descriptor(iterable):
    out = 'Diviner pipes descriptor file\n'
    for col in iterable:
        out += "'{0}' '{0}'\n".format(col)
    out += "'end' 'end'\n"
    return out
```

```
obs = fu.DivObs('2011120100')
```

```
l1a = obs.get_l1a()
```

```
print cols_to_descriptor(l1a.columns[:5])
```
