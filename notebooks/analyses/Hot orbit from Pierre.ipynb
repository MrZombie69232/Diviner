```python
from diviner import file_utils as fu
from diviner import plot_utils as pu
from diviner import calib
import pandas as pd
```

```python
def get_rdr2(tstr):
    df = fu.open_and_accumulate(tstr)
    rdr2 = calib.Calibrator(df)
    rdr2.calibrate()
    df['b3_11_offsets'] = rdr2.offsets_interp.b3_11
    return df, rdr2
```

```python
def compare_offsets_and_temps(df, tstr, temps=None):
    if temps is None:
        temps = 'oba_2_temp fpb_temp baffle_b_temp bb_2_temp'.split()
    df['offsets_shifted'] = df.b3_11_offsets.shift(8000)
    df['b3_11_offsets offsets_shifted'.split()+temps].\
            plot(style='.', markersize=7,
                 secondary_y=temps,
                 title=tstr)
    for i in range(1, df.space_block_labels.max()):
        mark_spaceview(df, i)
    for i in range(1, df.bb_block_labels.max()):
        mark_bbview(df, i)
```

```python
def mark_spaceview(df, label_no):
    t1 = df[df.space_block_labels==label_no].index[0]
    t2 = df[df.space_block_labels==label_no].index[-1]
    axvspan(*pd.to_datetime([t1,t2]), facecolor='blue', alpha=0.3)
def mark_bbview(df, label_no):
    t1 = df[df.bb_block_labels==label_no].index[0]
    t2 = df[df.bb_block_labels==label_no].index[-1]
    axvspan(*pd.to_datetime([t1,t2]), facecolor='red', alpha=0.3)
```

```python
# tstr = '2010040406'
tstr = '2010041806'
df, rdr2 = get_rdr2(tstr)
```

```python
from scipy.interpolate import UnivariateSpline as Spline
def interpolate_col(df, col):
    all_times = df.index.values.astype('float')
    col_times = df[col][df[col].notnull()].index.astype('float')
    spline = Spline(col_times, df[col], s=0.0, k=1)
    df[col+'_interp'] = spline(all_times)
```
