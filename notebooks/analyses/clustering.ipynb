```python
df = pd.read_excel('/Users/maye/data/diviner/madeup_CF_parameter_plot_simplified.xlsx', 0)
```

```python
df.columns = 'center asym CF'.split()
```

```python
df
```

```python
from sklearn.cluster import KMeans
```

```python
estimator = KMeans(init='k-means++', n_clusters=4)
```

```python
result = estimator.fit(df)
```

```python
result.labels_
```

```python
df['labels'] = result.labels_
```

```python
result.cluster_centers_
```

```python
fig = figure()
ax = fig.add_subplot(111, projection='3d')
colors = list('cymk')
for c, l in zip(colors, xrange(4)):
    to_plot = df[df.labels == l]
    ax.scatter(to_plot.asym, to_plot.center, to_plot.CF, c=c)
for center, col in zip(result.cluster_centers_, colors):
    ax.scatter(center[1], center[0], center[2], marker='^', c=col)
```

```python
from sklearn.cluster import AffinityPropagation
```

```python
af = AffinityPropagation().fit(df['center asym CF'.split()])
```

```python
af.cluster_centers_indices_
```

```python
af.labels_
```

```python
AffinityPropagation().fit_predict(df['center asym CF'.split()])
```

```python
fig = figure()
ax = fig.add_subplot(111, projection='3d')
colors = cycle('bgrcmyk')
for k, col in zip(xrange(len(af.cluster_centers_indices_)), colors):
    to_plot = df[af.labels_ == k]
    ax.scatter(to_plot.asym, to_plot.center, to_plot.CF, c=col)
```
