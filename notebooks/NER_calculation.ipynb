## Detrending how
As said before, one can see in above's plot that the blackbody has a significant gradient in the data.
I thought quite a bit abouth how to deal with this best. Tim mentions that he uses $$Counts(n)-Counts(n-16)$$ as a detrending scheme. As I was not sure which of many methods are best, I studied it below.

For a start, let's see how the values change if I skip Diviner's standard skipping of 16 samples after the telescope stopped moving:

```python
print sv[16:].std()
print bb[16:].std()
```

```python
print sv[16:].std()/sv.std()
print bb[16:].std()/bb.std()
```

Especially for the blackbody, this is a significant change of 20%. But what if the thermal gradient is not finished yet after 16 samples? I study here how the determined STD changes with increasing number of samples skipped and plot the results.

```python
sv_stds = []
bb_stds = []
for i in range(80):
    sv_stds.append(sv[i:].std())
    bb_stds.append(bb[i:].std())
sv_stds = np.array(sv_stds)
bb_stds = np.array(bb_stds)
```

```python
plot(range(80), sv_stds, 'b', range(80), bb_stds, 'r')
legend(['SV_STD', 'BB_STD'],loc='best')
```

The interesting thing here to note that it looks like the STD does depend on how it is taken. 

Let's look at the mean values between index 30-50

```python
print 'sv',sv_stds[30:50].mean()
print 'bb',bb_stds[30:50].mean()
sv_stds[30:50].mean() / bb_stds[30:50].mean()
```

Let's compare the number of 2.28 with other methods of determining the STD.

#### scipy.signal.detrend
First, I will be using a detrending routine from signal processing libraries. This works by removing a least-square linear fit from trending data.

```python
from scipy import signal
detrended = signal.detrend(bb) # detrending the BB view only
detrended.std() 
```

This value is very close to above's estimation. 

#### Tim's method
Let's now use Tim's method of making a manual differential:

```python
bb_differential = []
for i in range(16, len(bb[16:])):
    bb_differential.append(bb[16:][i] - bb[16:][i-16])
bb_differential = np.array(bb_differential)
bb_differential.std()
```

This value is quite higher than the other values determined before. Let's look at what the detrendings actually do to the data for further judgement.

#### Effect of different detrending methods on signal
For cross-check let's look at how the blackbody data actually looks like after this methods. For comparison, I subtract the mean value of the blackbody data, as the other detrending methods do that as well.
1. (red, solid) Just subtracting mean from the original (and 16 sample skipped) signal
2. (blue, dashed) Tim's method
3. (green, dot-dash) Detrending via signal processing library

```python
figure(figsize=(9,6))
plot(range(16, len(bb)), bb[16:]-bb[16:].mean(), '-r',
     label='orig - mean(orig),STD:{0:.2f}'.format((bb[16:]-bb[16:].mean()).std()))
plot(range(16, len(bb[16:])), bb_differential,'--b',
     label="Tim's method, STD:{0:.2f}".format(bb_differential.std()))
plot(detrended, '-.g', 
     label="detrending, STD:{0:.2f},{1:.2f}".format(detrended.std(),detrended[16:].std()), lw=2)
legend(loc='best')
```

Because the detrending removes a downwards trend throughout all the BB data, while it's not really trending anymore at the end, the detrended data points up at the end. It pulls down the front bit though and stays throughout pretty close to the shape of the original data.

However, Tim's method creates spikes that don't seem to be in the original data's STD, maybe I'm doing it wrong?

Both the simple method of skipping the first 16 samples and the detrending provide a value close to my above study on determining how many values one would have to skip before having found a good value for STD.

I therefore am confident that a value of 2.28 for the STD is a good value, if possibly still slighly conservative.

### Actual NER and NEDT calculation
Above I called my calibration routine where I calculate offset value and gains for the provided block of data.

Now I'm storing these values for detector b3_11 from the 8th position of the rdr2 objects offset and gain container:

```python
offset = rdr2.offsets[det].iloc[8]
gain = rdr2.gains[det].iloc[8]
```

As a reminder, I will be using this formula for NER:

$$ NER = \frac{SDT\cdot (R1-R2)}{(C1-C2)} $$
with 
* SDT the above determined standard deviation,
* `C1/2` the counts for two signals, and
* `R1/2` the calculated radiance signals for these counts.

For convenience, I will define a short function for it:

```python
def calc_NER(SDT, C1, C2, gain, offset):
    R1 = (C1 - offset) * gain
    R2 = (C2 - offset) * gain
    print 'R1: {}\nR2: {}'.format(R1, R2)
    return SDT*(R1-R2)/(C1-C2)
```

Note that I am working in normalized radiance. 

Choosing 
* the above determined conservative 2.28 as SDT,
* average of the spaceview as C1 (skipping 16)
* average of the blackbody view as C2 (skipping 16)

```python
c1 = sv[16:].mean()
c2 = bb[16:].mean()
# offset = c1
print c1, c2, offset
```

### NEdT calculation
As advised by Tim, I need to determine the gradient dR/dT for a more precise NEdT calculation.

I will be using the same $R_{BB}$ converter that I use during calibration, create a temperature array, feed it to the converter with the argument 9 for channel 9 and apply the gradient generator on the result, using the delta-Temps as argument for the differential.

Note that for consistency, this also has to stay in normalized radiance.

```python
rbbtable = calib.RBBTable()
temps = linspace(0, 400, 401)
radiances = rbbtable.get_radiance(temps, 9)
dR_dT = np.gradient(radiances, np.diff(temps)[0])
bucket = pd.DataFrame(dR_dT,index = temps, columns=['dR/dT'])
bucket['R_norm'] = radiances
bucket.plot(secondary_y='R_norm')
```

As the formula for NEdT is

$$ NEDT = \frac{NER}{dR/dT\left(T\right)} $$

with the dR/dT a function of T, one can just multiply the reciprol of above dR/dT curve with NER to create a temperature dependent NEdT curve. I'm also taking the absolute value of the result, because of our system where gains are negative, the NEdT also became negative

```python
bucket['NEdT'] = np.abs(NER / bucket['dR/dT'])
```

```python
# Zooming in
bucket.NEdT.plot(xlim=(10, 350), ylim=(0,20))
title("NEdT(T)")
ylabel("NEdT [K]")
xlabel("Temperature [K]")
savefig("/Users/maye/Dropbox/DDocuments/DIVINER/errors/nedt_0_to_150.png")
```
