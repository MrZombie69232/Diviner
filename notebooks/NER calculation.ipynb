Tim Schofield:
> If you have two known signal levels R1 and R2, and they give mean counts C1 and C2, then

> $$ NER = \frac{SD*(R1-R2)}{(C1-C2)} $$ 

> where SD is the standard deviation (one could use the mean of SD1 and SD2 for the two signal levels. Then,

> $$ NEdT = \frac{NER}{\frac{dR}{dT}} $$

> where dR/dT is derived from the lookup table of integrated radiance vs temperature

```python
from scipy import signal
```

```python
bb_differential = []
for i in range(16, len(bb)):
    bb_differential.append(bb[i] - bb[i-16])
bb_differential = np.array(bb_differential)
```

```python
detrended = signal.detrend(bb)
```

```python
plot(bb-bb.mean(), '-r', label='orig - mean(orig),STD:{0:.1f}'.format((bb-bb.mean()).std()))
plot(range(16, len(bb)), bb_differential,'--b', label="Tim's method, STD:{0:.1f}".format(bb_differential.std()))
plot(detrended, '-.g', label="detrending, STD:{0:.1f}".format(detrended.std()), lw=2)
legend(loc='best')
```
